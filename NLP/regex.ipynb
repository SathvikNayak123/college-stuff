```python
How to Find or Validate an Email Address
^[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}$

# has only a-z before @
# after @ has atleast one A-Z or 0-9 or . or -
# after . has A-Z with min length 2 and max 4

msis@MANIPAL2024.EDU
```
