```
X[0,:] # (2, 2, 3) first row as a vector =>rank 1 tensor
X[0,:].shape # (3,)
X[:,1] # (2,-1) first col as a vector =>rank 1 tensor
X[:,1].shape # (2,)

X[0,1:] # array([2, 4])
X[0,:-1] # array([1, 2])
```

```
array2=np.array([76,76,86,74,88,72,87])
array2 >=80 # array([False, False,  True, False,  True, False,  True])

# Boolean indexing of numpy arrays
array2[array2>=80] # array([86, 88, 87])

#manual shuffling
indexs=[6,4,2,3,1,0,6]
array2[indexs] # array([87, 88, 86, 74, 76, 76, 87])
```

```
X=np.array([
    [1,2,4],
    [2,-1,3]
])

X.sum()
np.sum(X)
np.sum(X,axis=1) # array([7, 4]) add row elems
np.sum(X,axis=0) # array([3, 1, 7]) add col elems

np.mean(X,axis=1) # array([2.33333333, 1.33333333])
np.max(X,axis=1) # array([4, 3])
```

```
#creaing array of certain shape
np.ones((1,3)) # array([[1., 1., 1.]])
np.ones((1,X.shape[1])) # same as above - shape value shld be given as tuple
```

```
# visualize the relationship between 2 contineous features
fig,ax = plt.subplots(1,figsize = (3,3))
ax.scatter(df['Oil'],df['Density'],color = 'green',s=15);
ax.set_xlabel('Oil Percentage')
ax.set_ylabel('Density')
ax.set_title('Density vs Oil Percentage');
```

```
bins = np.arange(min(df['Oil']),max(df['Oil']),0.5)
```

```
# visualize a single contineous feature: Oil
# Histogram
fig,ax = plt.subplots(1,figsize = (5,5))
ax.hist(df['Oil'],bins);
ax.set_xlabel('Oil Percentage')
ax.set_ylabel('Count');
```

```
oil=df['Oil']
oil_std=(oil - np.mean(oil)) / (np.std(oil))
plt.hist(oil_std,bins=10);
```

```
def calc_std(x,y):
  a=np.mean(x) - y*np.std(x)
  b=np.mean(x) + y*np.std(x)
  print(np.mean((x>=a) & (x<=b)))
```

```
calc_std(df['Oil'],1.75)
```

```
df.index[np.abs(oil_std)>1.75]
```

```
calc_std(df['Oil'],1)
calc_std(df['Oil'],2)
calc_std(df['Oil'],3)
```

```
np.mean(np.abs(df['Oil']-np.mean(df['Oil'])) <= 1*np.std(df['Oil']))
```
